### Cervical Cytology Categorization

#### Categorized Cells (Main Folder)
This folder contains all categorized cell images for cytological analysis, organized into three categories: Normal Cells, Reactive or Reparative Changes, and Pre-cancerous Lesions.

#### Normal Cells
Healthy cells from the cervix, showing no abnormality.
- **Intermediate Cells**: Mid-layer cells of normal squamous epithelium.
- **Superficial Cells**: Mature cells from the outer cervical layer.
- **Parabasal Cells**: Deeper cells, normal in certain conditions.
- **Basal Cells**: Deepest layer of squamous epithelium.

#### Reactive or Reparative Changes
Benign changes in response to inflammation or irritation.
- **Squamous Metaplasia**: Replacement of glandular cells by squamous cells, a normal physiological process.

#### Pre-cancerous Lesions
Abnormal cells with potential to develop into cancer.
- **Koilocytes**: Cells showing HPV infection, associated with low-grade lesions.
- **Dysplastic Cells**: Abnormal growth, indicating mild to severe pre-cancerous conditions.


```python
import os

# Define the base paths for each category
base_path = "C:/Users/PC/OneDrive/Escritorio/VPH/XMAYUS/OneDrive_2024-09-18/Proyecto vph/Categorized Cells"
categories = ["Normal Cells", "Pre-cancerous Lesions", "Reactive or Reparative Changes"]

# Initialize a dictionary to store the count of images per category
image_count = {category: 0 for category in categories}
formats_count = {}

# Process images in each category and delete non-JPG/PNG images
for category in categories:
    category_path = os.path.join(base_path, category)
    for root, _, files in os.walk(category_path):
        for file in files:
            file_path = os.path.join(root, file)
            file_extension = os.path.splitext(file)[1].lower()
            # Initialize count for this format if not already present
            if file_extension not in formats_count:
                formats_count[file_extension] = 0
            formats_count[file_extension] += 1
            # Delete images not in .jpg or .png format
            if file_extension not in ['.jpg', '.jpeg', '.png']:
                os.remove(file_path)
                formats_count[file_extension] -= 1
            else:
                # Count images in the desired formats
                image_count[category] += 1

# Output the count of different formats after processing
formats_count, image_count
```
